```python
sns.boxplot(x='sex', y='age', data=titanic, hue='survived')
plt.title("Age Distribution by Sex and Survival Status")
plt.xlabel("Sex")
plt.ylabel("Age")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.show()
```

## The Violin Plot
Violin Plot: Combines boxplot + KDE (kernel density estimate).


```python
sns.violinplot(x='sex', y='age', data=titanic)
plt.title("Age Distribution by Sex (Violin Plot)")
plt.xlabel("Sex")
plt.ylabel("Age")
plt.show()
```

```python
sns.violinplot(x='sex', y='age', data=titanic, hue='survived', split=True)
plt.title("Age Distribution by Sex and Survival (Violin Plot)")
plt.xlabel("Sex")
plt.ylabel("Age")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.show()
```

## The Strip Plot
A scatterplot for categorical data.
Jitter: Adds small noise to avoid overlapping.


```python
# Strip plot
sns.stripplot(x='sex', y='age', data=titanic, jitter=True)
plt.title("Age Distribution by Sex (Strip Plot)")
plt.xlabel("Sex")
plt.ylabel("Age")
plt.show()
```

```python
# Strip plot with hue
#Jitter: Adds small noise to avoid overlapping.
# Dodge: Separates the hue values.

sns.stripplot(x='sex', y='age', data=titanic, jitter=True, hue='survived', dodge=True)
plt.title("Age Distribution by Sex and Survival (Strip Plot)")
plt.xlabel("Sex")
plt.ylabel("Age")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.show()
```

## The Swarm Plot
Swarm Plot: Improved version of stripplot â€” avoids overlapping.

```python
sns.swarmplot(x='sex', y='age', data=titanic)
plt.title("Age Distribution by Sex (Swarm Plot)")
plt.xlabel("Sex")
plt.ylabel("Age")
plt.show()
```

```python
sns.swarmplot(x='sex', y='age', data=titanic, hue='survived', dodge=True)
plt.title("Age Distribution by Sex and Survival (Swarm Plot)")
plt.xlabel("Sex")
plt.ylabel("Age")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.show()
```
