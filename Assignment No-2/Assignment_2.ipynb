### 2. Detect outlier using z-Score

```python
from scipy import stats


# Select only numerical columns
#This filters the DataFrame to include only columns with numeric data 
# (e.g., marks, percentage). Non-numeric columns like 'Name', 'Result', or 
# 'Term' (if it's object/string type) are excluded.
numeric_df = df.select_dtypes(include=[np.number])

# Calculate z-scores
# stats.zscore() computes the z-score for each numeric value (how far it is from the column mean, in terms of standard deviation).
# np.abs() takes the absolute value so we don’t care if the outlier is above or below the mean.
# A z-score > 3 usually means the value is an outlier.
z_scores = np.abs(stats.zscore(numeric_df))

# Define a threshold
threshold = 3

# Find outliers
# This gives a boolean DataFrame where:
# True = outlier
# False = normal
outliers = (z_scores > threshold)

# Get rows with any outlier
# .any(axis=1) checks if any column in a row has an outlier.
# np.where(...)[0] gets the row indices where outliers exist.
outlier_rows = np.where(outliers.any(axis=1))[0]

# Print rows with outliers
# Now you’re printing only those rows from the original DataFrame that have at least one extreme value.
df_outliers = df.iloc[outlier_rows]
print(df_outliers)


# Output Explanation
# Let’s look at one row:

# plaintext
# Copy code
# Rollno: 11, Subject1_marks: 490.0
# A subject score like 490 is very unusual, so it gets a high z-score, marking it as an outlier.

# Another row:

# plaintext
# Copy code
# Rollno: 14, Subject1_marks: 1.0, Subject2_marks: 1.0, Subject3_marks: 1.0
# Extremely low values across subjects make this row stand out — again flagged as an outlier.
```

### 3.Detecting outliers using Inter Quantile Range(IQR):

```python
Q1 = df['Percentage'].quantile(0.25)
Q3 = df['Percentage'].quantile(0.75)
IQR = Q3 - Q1
Lower_limit = Q1 - 1.5 * IQR
Upper_limit = Q3 + 1.5 * IQR
print(f'Q1 = {Q1}, Q3 = {Q3}, IQR = {IQR}, Lower_limit = {Lower_limit}, Upper_limit = {Upper_limit}')

# The Interquartile Range (IQR) is a measure of statistical spread, showing the range of the middle 50% of your data.

# Definition:
# IQR=Q3−Q1
# Where:

# Q1 (1st quartile) = 25th percentile (the value below which 25% of the data fall)

# Q3 (3rd quartile) = 75th percentile (the value below which 75% of the data fall)

# Meaning:
# IQR tells you how spread out the central half of your data is. It's useful for identifying outliers.
```

```python
df[(df['Percentage'] < Lower_limit) | (df['Percentage'] > Upper_limit)] # outlier data
```

### The variables with skewness < -1 are highly negatively skewed.
### The variables with 0.5 < skewness < 1 such as moderately positively skewed.
### The variables with -0.5 < skewness < -1 such as stroke are moderately negatively skewed.
### And, the variables with -0.5 < skewness < 0.5 are symmetric i.e normally distributed such as symboling, carheight, boreration, peakrpm, highwaympg.
